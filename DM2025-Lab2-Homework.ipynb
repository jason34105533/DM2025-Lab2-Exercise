### 1.1 Preprocessing Steps

For this competition, the raw data was distributed across three files: `data_identification.csv`, `emotion.csv`, and `final_posts.json`. My preprocessing pipeline involved the following steps:

1. **Data Integration:**

   - I parsed the JSON structure of `final_posts.json` to extract the text content and hashtags.
   - I merged the three datasets on the unique `id` field to create a single master dataframe containing the text, split status (train/test), and ground truth emotion labels.

2. Text Cleaning:

   To reduce noise and standardize the input for the model, I applied a cleaning function to the raw text:

   - **Lowercasing:** Converted all text to lowercase to ensure consistency.
   - **URL Removal:** Used regex to strip out HTTP/HTTPS links, as they generally do not carry emotional sentiment.
   - **User Mention Removal:** Removed `@username` patterns to prevent the model from overfitting on specific user handles.
   - **Whitespace Handling:** Removed extra spaces and stripped leading/trailing whitespace.

3. **Data splitting:**

   - Finally, the data was separated back into `train_df` (labeled) and `test_df` (unlabeled) based on the provided `split` column.

### 1.2 Feature Engineering Steps

To prepare the text data for the BERT model, I performed the following feature engineering tasks:

1. **Label Encoding:**
   - The string labels (`joy`, `anger`, `surprise`, `sadness`, `fear`, `disgust`) were mapped to integers (0-5) to be compatible with the model's loss function.
   - *Mapping:* `{'joy': 0, 'anger': 1, 'surprise': 2, 'sadness': 3, 'fear': 4, 'disgust': 5}` (Adjust this list if your specific run generated a different order).
2. **Stratified Validation Split:**
   - I split the training data into a training set (80%) and a validation set (20%).
   - **Crucial Step:** I used **stratified splitting** to ensure that the distribution of emotions in the validation set exactly matched the training set. This is important because some emotions (like `fear` or `disgust`) are underrepresented in the dataset.
3. **BERT Tokenization:**
   - I used the `BertTokenizer` from the `bert-base-uncased` model.
   - **Truncation & Padding:** sequences were truncated to a maximum length of **128 tokens** and padded to ensure uniform tensor shapes within batches. This length was chosen to cover the majority of tweets without wasting computational resources on empty padding.

### 1.3 Explanation of Your Model

I utilized a **Transfer Learning** approach by fine-tuning a pre-trained BERT model.

- **Model Architecture:**
  - I used `BertForSequenceClassification` initialized with the `bert-base-uncased` weights.
  - This model consists of a 12-layer Transformer encoder. The output of the final layer's `[CLS]` token is fed into a linear classification layer that projects the hidden state to the 6 emotion categories.
- **Training Strategy:**
  - I used the Hugging Face `Trainer` API for the training loop.
  - **Hyperparameters:**
    - **Epochs:** 3
    - **Batch Size:** 16 for training, 32 for evaluation.
    - **Optimization:** AdamW optimizer with weight decay (0.01) to prevent overfitting.
    - **Learning Rate:** A linear warmup scheduler was used for the first 500 steps to stabilize early training.
- **Metric:**
  - The model was evaluated using **Weighted F1-Score** and **Accuracy**. Weighted F1 was prioritized to account for the class imbalance in the dataset.

## 1. Preprocessing Steps

```python
### Add the code related to the preprocessing steps in cells inside this section

# ./dm-lab-2-private-competition/data_identification.csv
'''
id,split
0x61fc95,test
0x35663e,train
0xc78afe,train
'''


# ./dm-lab-2-private-competition/emotion.csv
'''
id,emotion
0x35663e,joy
0xc78afe,fear
0x90089c,joy
'''

# ./dm-lab-2-private-competition/final_post.json
'''
[
  {
    "root": {
      "_type": "post",
      "_source": {
        "post": {
          "post_id": "0x61fc95",
          "text": "We got the ranch, loaded our guns and sat up till sunrise.",
          "hashtags": []
        }
      }
    }
  },
  {
    "root": {
      "_type": "post",
      "_source": {
        "post": {
          "post_id": "0x35663e",
          "text": "I bet there is an army of married couples who did the same exact thing.",
          "hashtags": []
        }
      }
    }
  },

'''


# Data Preprocessing
import pandas as pd
import json
from pathlib import Path
from sklearn.model_selection import train_test_split
import os

# Load data
data_dir = Path('./dm-lab-2-private-competition')
identification_df = pd.read_csv(data_dir / 'data_identification.csv')
emotion_df = pd.read_csv(data_dir / 'emotion.csv')
with open(data_dir / 'final_posts.json', 'r') as f:
    posts_data = json.load(f)
posts_df = pd.DataFrame([item['root']['_source']['post'] for item in posts_data])
posts_df.rename(columns={'post_id': 'id'}, inplace=True)

# Merge dataframes
merged_df = identification_df.merge(emotion_df, on='id', how='left').merge(posts_df, on='id', how='left')


# Clean the text data by removing extra spaces, url, and @username and converting to lowercase
def clean_text(text):
    import re
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)  # Remove URLs
    text = re.sub(r'@\w+', '', text)  # Remove @username
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

merged_df['text'] = merged_df['text'].apply(clean_text)

# Split into train and test sets
train_df = merged_df[merged_df['split'] == 'train'].reset_index(drop=True)
test_df = merged_df[merged_df['split'] == 'test'].reset_index(drop=True)

# Save processed data
output_dir = Path('./processed_data')
os.makedirs(output_dir, exist_ok=True)
train_df.to_csv(output_dir / 'train_data.csv', index=False)
test_df.to_csv(output_dir / 'test_data.csv', index=False)

print("Data preprocessing completed. Processed files are saved in './processed_data' directory.")

print("Train Data:")
print(train_df.head())
print("Test Data:")
print(test_df.head())
```

## 2. Feature Engineering Steps

```python
import torch
from transformers import BertTokenizer
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset

# 1. Define Label Mapping
# Based on the distribution in your logs: joy, anger, surprise, sadness, fear, disgust
possible_labels = train_df['emotion'].unique()
label_map = {label: i for i, label in enumerate(possible_labels)}
id2label = {i: label for label, i in label_map.items()}

print(f"Label Mapping: {label_map}")
print(f"Number of classes: {len(label_map)}")

# 2. Split Training Data into Train and Validation
# It is best practice to have a validation set to monitor accuracy during training
X_train, X_val, y_train, y_val = train_test_split(
    train_df['text'].tolist(),
    train_df['emotion'].map(label_map).tolist(),
    test_size=0.2,
    random_state=42,
    stratify=train_df['emotion']
)

# 3. Load Tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# 4. Tokenization Function
def tokenize_data(texts, labels=None):
    encodings = tokenizer(texts, truncation=True, padding=True, max_length=128)
    if labels:
        # Create a list of dictionaries for the Dataset
        dataset = []
        for i in range(len(texts)):
            item = {key: torch.tensor(val[i]) for key, val in encodings.items()}
            item['labels'] = torch.tensor(labels[i])
            dataset.append(item)
        return dataset
    else:
        # For test data (no labels)
        dataset = []
        for i in range(len(texts)):
            item = {key: torch.tensor(val[i]) for key, val in encodings.items()}
            dataset.append(item)
        return dataset

# 5. Create Datasets
train_dataset = tokenize_data(X_train, y_train)
val_dataset = tokenize_data(X_val, y_val)

# Prepare Test Data
test_texts = test_df['text'].tolist()
test_dataset = tokenize_data(test_texts)

print("Feature Engineering Completed. Datasets ready.")
```

```python
# plot the distribution of labels in the training set
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

label_counts = train_df['emotion'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=label_counts.index, y=label_counts.values, palette='viridis')
plt.title('Distribution of Emotion Labels in Training Set')
plt.xlabel('Emotion')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

## 3. Model Implementation Steps

```python
from transformers import BertForSequenceClassification, Trainer, TrainingArguments
import numpy as np
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

# 1. Compute Metrics Function
# This allows us to see accuracy/F1 score during training
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='weighted')
    acc = accuracy_score(labels, preds)
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }

# 2. Load Pretrained Model
# CRITICAL: num_labels must equal the number of emotion categories (6 in your case)
model = BertForSequenceClassification.from_pretrained(
    'bert-base-uncased',
    num_labels=len(label_map)
)

# 3. Define Training Arguments
# Adjust batch_size depending on your GPU VRAM (try 16 or 8 if you get OOM errors)
training_args = TrainingArguments(
    output_dir='./results',          # Output directory
    num_train_epochs=3,              # Total number of training epochs
    per_device_train_batch_size=16,  # Batch size per device during training
    per_device_eval_batch_size=32,   # Batch size for evaluation
    warmup_steps=500,                # Number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # Strength of weight decay
    logging_dir='./logs',            # Directory for storing logs
    logging_steps=100,
    eval_strategy="epoch",     # Evaluate every epoch
    save_strategy="epoch",           # Save checkpoint every epoch
    load_best_model_at_end=True,     # Load the best model when finished training
    metric_for_best_model="accuracy"
)

# 4. Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
)

# 5. Train the Model
print("Starting Training...")
trainer.train()

# 6. Predict on Test Set
print("Generating Predictions...")
predictions = trainer.predict(test_dataset)
preds_indices = np.argmax(predictions.predictions, axis=-1)

# 7. Map Predictions back to Label Names
pred_labels = [id2label[idx] for idx in preds_indices]

# 8. Create Submission File
# Merging prediction back to the test dataframe structure
submission = test_df[['id']].copy()
submission['emotion'] = pred_labels

# Save to CSV
submission.to_csv('submission.csv', index=False)
print("Submission file 'submission.csv' created successfully!")
print(submission.head())
```

## 4. Visualizations and Error Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import pandas as pd

# --- 4.1 Training History Visualization ---
# Extract logs from the trainer
history = trainer.state.log_history

# Separate training loss and evaluation metrics
train_loss = []
eval_loss = []
eval_acc = []
steps = []
eval_steps = []

for entry in history:
    if 'loss' in entry:
        train_loss.append(entry['loss'])
        steps.append(entry['step'])
    if 'eval_loss' in entry:
        eval_loss.append(entry['eval_loss'])
        eval_acc.append(entry['eval_accuracy'])
        eval_steps.append(entry['step'])

# Plotting Loss Curves
plt.figure(figsize=(12, 5))

# Subplot 1: Loss
plt.subplot(1, 2, 1)
plt.plot(steps, train_loss, label='Training Loss', color='blue', alpha=0.6)
plt.plot(eval_steps, eval_loss, label='Validation Loss', color='red', marker='o')
plt.title('Training and Validation Loss')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Subplot 2: Accuracy
plt.subplot(1, 2, 2)
plt.plot(eval_steps, eval_acc, label='Validation Accuracy', color='green', marker='s')
plt.title('Validation Accuracy over Time')
plt.xlabel('Steps')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# --- 4.2 Confusion Matrix on Validation Set ---
print("Generating Confusion Matrix on Validation Set...")

# Get predictions on the validation dataset
val_predictions = trainer.predict(val_dataset)
val_preds_indices = np.argmax(val_predictions.predictions, axis=-1)
val_true_indices = val_predictions.label_ids

# Create Confusion Matrix
cm = confusion_matrix(val_true_indices, val_preds_indices)

# Normalize the confusion matrix (helps to see percentages)
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

# Get label names in correct order
label_names = [id2label[i] for i in range(len(id2label))]

# Plot Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues',
            xticklabels=label_names, yticklabels=label_names)
plt.title('Normalized Confusion Matrix (Validation Set)')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()

# --- 4.3 Error Analysis ---
# Show some actual examples where the model failed
print("\n--- Error Analysis Examples ---")
val_texts = X_val  # Original text from the split step
mistakes = []

for i in range(len(val_true_indices)):
    if val_true_indices[i] != val_preds_indices[i]:
        mistakes.append({
            'text': val_texts[i],
            'true': id2label[val_true_indices[i]],
            'predicted': id2label[val_preds_indices[i]]
        })
        if len(mistakes) >= 5: # Only show first 5 mistakes
            break

for m in mistakes:
    print(f"Text: {m['text']}")
    print(f"True: {m['true']} | Predicted: {m['predicted']}")
    print("-" * 50)
```
